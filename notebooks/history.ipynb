```python
import pandas as pd
import numpy as np
import re
```

```python
file_path = "../data/landing/Moving annual rent by suburb - March quarter 2023 copy.xlsx"
df_1_bedroom_flat_cleaned = pd.read_excel(file_path, sheet_name='1 bedroom flat', header=[0, 1])
df_1_bedroom_flat_cleaned.columns = [
    f'{col1} {col2}'.strip() if pd.notna(col1) and pd.notna(col2) else col1 or col2 
    for col1, col2 in df_1_bedroom_flat_cleaned.columns
]
df_1_bedroom_flat_cleaned.drop([0], inplace=True)
df_1_bedroom_flat_cleaned.reset_index(drop=True, inplace=True)
```

```python
df_1_bedroom_flat_cleaned
```

```python
df_1_bedroom_flat_cleaned.columns.values[0] = 'Suburb'
df_1_bedroom_flat_cleaned.columns.values[1] = 'Region'
df_1_bedroom_flat_cleaned.columns
```

```python
print(df_1_bedroom_flat_cleaned.columns)
df_1_bedroom_flat_cleaned.columns = df_1_bedroom_flat_cleaned.columns.str.strip()
```

```python
count_col = [col for col in df_1_bedroom_flat_cleaned.columns if '.1' not in col]
print("Columns without '.1':", count_col)
count_df = df_1_bedroom_flat_cleaned[count_col]
count_df.reset_index(drop=True, inplace=True)
```

```python
count_df
```

```python
median_df = df_1_bedroom_flat_cleaned[['Suburb', 'Region']+[col for col in df_1_bedroom_flat_cleaned.columns if '.1' in col]]
median_df.columns = [col.replace('.1', '').strip() for col in median_df.columns]
median_df.reset_index(drop=True, inplace=True)
```

```python
median_df
```

```python
def extract_date(col_name):
    match = re.findall(r'Lease commenced in year ending  (\w{3} \d{4})', col_name)
    return match[0] if match else col_name
```

```python
count_df.columns = [extract_date(col) for col in count_df.columns]
median_df.columns = [extract_date(col.replace('.1', '')) for col in median_df.columns]
```

```python
count_df_calc = count_df.drop('Suburb', axis=1)
count_df_calc
```

```python
count_df_growth =count_df_calc.copy()
growth_df = pd.DataFrame()
growth_df['Region'] = count_df_growth['Region']
for i in range(1, len(count_df_growth.columns) - 1):
    now = pd.to_numeric(count_df_growth.iloc[:, i], errors='coerce')
    later = pd.to_numeric(count_df_growth.iloc[:, i + 1], errors='coerce')
    growth = np.where((now > 0) & (later.notna()), ((later - now) / now) * 100, np.nan)
    growth_df[f'Growth_Rate_{count_df_growth.columns[i]}_to_{count_df_growth.columns[i+1]}'] = growth
```

```python
growth_df
```

```python
median_df_calc = median_df.drop('Suburb', axis=1)
median_df_growth =median_df_calc.copy()
mgrowth_df = pd.DataFrame()
mgrowth_df['Region'] = median_df_growth['Region']
for i in range(1, len(median_df_growth.columns) - 1):
    now = pd.to_numeric(median_df_growth.iloc[:, i], errors='coerce')
    later = pd.to_numeric(median_df_growth.iloc[:, i + 1], errors='coerce')
    growth = np.where((now > 0) & (later.notna()), ((later - now) / now) * 100, np.nan)
    mgrowth_df[f'Growth_Rate_{median_df_growth.columns[i]}_to_{median_df_growth.columns[i+1]}'] = growth
```

```python
mgrowth_df
```

```python
import plotly.express as px
import pandas as pd

growth_df_long = pd.melt(growth_df, id_vars=['Region'], 
                         var_name='Time_Period', value_name='Growth_Rate')

fig = px.line(growth_df_long, 
              x='Time_Period', y='Growth_Rate', color='Region', 
              title='Growth Rate Over Time by Region')

fig.update_layout(
    xaxis_title='Time Period',
    yaxis_title='Growth Rate (%)',
    hovermode='x unified',
    legend_title='Region',
    xaxis=dict(tickangle=45),  
)

fig.show()
```
